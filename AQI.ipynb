1. List the top 5 and bottom 5 areas with highest average AQI. (Consider areas which contains data from last 6 months: December 2024 to May 2025)

```python
import pandas as pd

try:
    # Convert 'date' column to datetime
    aqi['date'] = pd.to_datetime(aqi['date'], format='%d-%m-%Y')

    # Filter data for the last 6 months (December 2024 to May 2025)
    start_date = '2024-12-01'
    end_date = '2025-05-31'
    filtered_df = aqi[(aqi['date'] >= start_date) & (aqi['date'] <= end_date)]

    # Calculate the average AQI for each area
    average_aqi = filtered_df.groupby('area')['aqi_value'].mean().round(2).reset_index()

    # Get the top 5 areas with the highest average AQI
    top_5_areas = average_aqi.sort_values(by='aqi_value', ascending=False).head(5)

    # Get the bottom 5 areas with the lowest average AQI
    bottom_5_areas = average_aqi.sort_values(by='aqi_value', ascending=True).head(5)

    # --- Print the Final Answer ---
    print("Based on the analysis of the data from December 2024 to May 2025:\n")

    print("--- Top 5 Areas with Highest Average AQI ---")
    print(top_5_areas.to_string(index=False))

    print("\n--- Bottom 5 Areas with Lowest Average AQI ---")
    print(bottom_5_areas.to_string(index=False))

except FileNotFoundError:
    print("Error: 'aqi.csv' not found. Please make sure the file is in the same directory as the script.")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
# --- Visualization ---

import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for the plots
sns.set_style("whitegrid")

# 1. Create the bar chart for the Top 5 Areas
plt.figure(figsize=(10, 6))
sns.barplot(x='aqi_value', y='area', data=top_5_areas, palette='Reds_r')
plt.title('Bottom 5 Least Polluted Areas (Avg AQI)\nDec 2024 – Apr 2025', fontsize=14)
plt.xlabel('Average AQI Value', fontsize=12)
plt.ylabel('Area', fontsize=12)
plt.show()

# 2. Create the bar chart for the Bottom 5 Areas
plt.figure(figsize=(10, 6))
sns.barplot(x='aqi_value', y='area', data=bottom_5_areas, palette='Greens_r')
plt.title('Bottom 5 Least Polluted Areas (Avg AQI)\nDec 2024 – Apr 2025', fontsize=14)
plt.xlabel('Average AQI Value', fontsize=12)
plt.ylabel('Area', fontsize=12)
plt.show()
```

2. List out top 2 and bottom 2 prominent pollutants for each state of southern India. 
(Consider data post covid: 2022 onwards) 

```python
aqi['state'].unique()
```

```python
import pandas as pd

try:
    # Convert 'date' column to datetime
    aqi['date'] = pd.to_datetime(aqi['date'], format='%d-%m-%Y', errors='coerce')

    # Filter data for the post-COVID period (2022 onwards)
    start_date = '2022-01-01'
    df_filtered = aqi[aqi['date'] >= start_date].copy()

    # Define the states of Southern India
    southern_states = [
        'Andhra Pradesh',
        'Karnataka',
        'Kerala',
        'Tamil Nadu',
        'Telangana',
        'Puducherry' # Union Territory
    ]

    # Filter the DataFrame to include only Southern Indian states
    df_southern = df_filtered[df_filtered['state'].isin(southern_states)]

    # Create an empty list to store the results
    results_list = []

    # Iterate through each southern state to perform the analysis
    for state in sorted(southern_states):
        state_df = df_southern[df_southern['state'] == state]

        if state_df.empty:
            # Add a row indicating no data
            results_list.append({
                'State': state,
                'Top 2 Pollutants': 'No Data Available',
                'Bottom 2 Pollutants': 'No Data Available'
            })
            continue

        # Split multi-pollutant strings and create a row for each pollutant
        pollutants_series = state_df['prominent_pollutants'].str.split(',').explode()
        pollutants_series = pollutants_series.str.strip()
        
        # Count the occurrences of each pollutant
        pollutant_counts = pollutants_series.value_counts()

        # Get the top 2 and bottom 2 pollutants
        top_2 = pollutant_counts.head(2)
        bottom_2 = pollutant_counts.tail(2)

        # Append the results for the state to our list
        results_list.append({
            'State': state,
            'Top 2 Pollutants': ', '.join(top_2.index.tolist()),
            'Bottom 2 Pollutants': ', '.join(bottom_2.index.tolist())
        })

    # Convert the list of results into a pandas DataFrame
    results_df = pd.DataFrame(results_list)

    # Print the final DataFrame as a clean table
    print("--- Top/Bottom Pollutants by State in Southern India (2022 onwards) ---")
    print(results_df.to_string(index=False))

except FileNotFoundError:
    print("Error: 'aqi.csv' not found. Please make sure the file is in the same directory.")
except Exception as e:
    print(f"An error occurred: {e}")
```

3. Does AQI improve on weekends vs weekdays in Indian metro cities (Delhi, 
Mumbai, Chennai, Kolkata, Bengaluru, Hyderabad, Ahmedabad, Pune)? 
(Consider data from last 1 year) 

```python
import pandas as pd
from dateutil.relativedelta import relativedelta

# Drop rows where date or AQI is missing
aqi = aqi.dropna(subset=['date', 'aqi_value'])

# Filter to last 1 year from latest date
latest_date = aqi['date'].max()
start_date = latest_date - relativedelta(years=1)

aqi_last_year = aqi[(aqi['date'] >= start_date) & (aqi['date'] <= latest_date)].copy()
```

```python
print(f'Latest_date: {latest_date} ')
print(f'Start_date: {start_date} ')
```

```python
# # Focus cities
metro_cities = ['Delhi', 'Mumbai', 'Chennai', 'Kolkata', 'Bengaluru', 'Hyderabad', 'Ahmedabad', 'Pune']

aqi_metros = aqi_last_year[aqi_last_year['area'].isin(metro_cities)].copy()
aqi_metros['day_type'] = aqi_metros['date'].dt.dayofweek.apply(lambda x: 'Weekend' if x >= 5 else 'Weekday')

# Group and compare
avg_aqi_by_day = (
    aqi_metros.groupby(['area', 'day_type'])['aqi_value']
    .mean()
    .round(2)
    .reset_index()
    .pivot(index='area', columns='day_type', values='aqi_value')
    .reset_index()
)

# Calculate improvement (positive = better on weekends)
avg_aqi_by_day['AQI_Improved_On_Weekend'] = avg_aqi_by_day['Weekday'] - avg_aqi_by_day['Weekend']

# Display result
print("Average AQI on Weekdays vs Weekends (Last 1 Year):\n")
avg_aqi_by_day
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Melt the dataframe
melted = avg_aqi_by_day.melt(id_vars='area', value_vars=['Weekday', 'Weekend'],
                             var_name='Day Type', value_name='Average AQI')

plt.figure(figsize=(12, 6))
ax = sns.barplot(data=melted, x='area', y='Average AQI', hue='Day Type', palette='Set2')

# Annotate each bar (loop through patches)
for bar in ax.patches:
    height = bar.get_height()
    ax.annotate(
        f'{height:.1f}',
        xy=(bar.get_x() + bar.get_width() / 2, height),
        xytext=(0, 4),  # offset above the bar
        textcoords='offset points',
        ha='center',
        va='bottom',
        fontsize=9,
        color='black'
    )

# Final touches
plt.title('Average AQI on Weekdays vs Weekends (Metro Cities, Last 1 Year)')
plt.xlabel('City')
plt.ylabel('Average AQI')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.legend(title='Day Type')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(x='area', y='AQI_Improved_On_Weekend', hue='area', data=avg_aqi_by_day, palette='coolwarm', dodge=False, legend=False)

plt.axhline(0, color='black', linewidth=0.8, linestyle='--')
plt.title('AQI Improvement on Weekends (Weekday - Weekend)', fontsize=14)
plt.ylabel('Improvement (+ = Better AQI on Weekend)')
plt.xlabel('City')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

4. Which months consistently show the worst air quality across Indian states — (Consider top 10 states with high distinct areas) 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

try:
    # Convert 'date' column to datetime
    aqi['date'] = pd.to_datetime(aqi['date'], format='%d-%m-%Y', errors='coerce')

    # --- Part 1: Find the top 10 states with the most distinct monitoring areas ---
    state_area_counts = aqi.groupby('state')['area'].nunique()
    top_10_states = state_area_counts.nlargest(10).index.tolist()
    
    print("--- Analysis will be based on the top 10 states with the most monitoring areas ---")
    print(top_10_states)
    print("-" * 70)

    # --- Part 2: Filter the main dataframe for these states ---
    df_top_10 = aqi[aqi['state'].isin(top_10_states)].copy()

    # Extract month name for grouping
    df_top_10['month'] = df_top_10['date'].dt.month
    df_top_10['month_name'] = df_top_10['date'].dt.strftime('%B')

    # --- Part 3: Calculate the average AQI for each month ---
    monthly_avg_aqi = df_top_10.groupby(['month', 'month_name'])['aqi_value'].mean().reset_index()
    
    # Sort by average AQI to find the worst months
    worst_months = monthly_avg_aqi.sort_values(by='aqi_value', ascending=False)

    print("\n--- Months Ranked by Worst Average AQI ---")
    print(worst_months[['month_name', 'aqi_value']].round(2).to_string(index=False))
    
    # --- Part 4: Visualize the results ---
    # Sort by month number for a proper chronological chart
    # Sort by aqi_value for a ranked chart where color matches height
    monthly_avg_aqi_sorted = monthly_avg_aqi.sort_values(by='aqi_value', ascending=False)
    monthly_avg_aqi_sorted['aqi_value'] = np.round(monthly_avg_aqi_sorted['aqi_value']).astype(int)

    plt.figure(figsize=(12, 7))
    ax = sns.barplot(x='month_name', y='aqi_value', data=monthly_avg_aqi_sorted, palette='OrRd_r')

    # Remove background grid and spines
    sns.despine()  # Removes top and right spines
    ax.set_facecolor('white')  # Plot background
    ax.figure.set_facecolor('white')  # Overall figure background
    plt.grid(False)  # Removes gridlines

    # Add data labels on each bar
    for i in ax.containers:
        ax.bar_label(i, fmt='%d', fontsize=10, label_type='edge', padding=3)

    plt.title("Worst Air Quality Months in India’s Top 10 States", fontsize=16)
    plt.xlabel('Month', fontsize=12)
    plt.ylabel('Average AQI Value', fontsize=12)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

except FileNotFoundError:
    print("Error: 'aqi.csv' not found. Please make sure the file is in the same directory.")
except Exception as e:
    print(f"An error occurred: {e}")
```

6. List the top two most reported disease illnesses in each state over the past three 
years, along with the corresponding average Air Quality Index (AQI) for that 
period. 

```python
vahan = pd.read_excel(r"C:\Users\08viv\Documents\Codebasics Challenge 16\cleaned data\Vahan.xlsx")
idsp = pd.read_excel(r"C:\Users\08viv\Documents\Codebasics Challenge 16\cleaned data\idsp.xlsx")
population = pd.read_excel(r"C:\Users\08viv\Documents\Codebasics Challenge 16\cleaned data\population.xlsx")
```

```python
import pandas as pd

try:
    # Convert date columns to datetime
    idsp['date'] = pd.to_datetime(idsp['outbreak_starting_date'], format='%d-%m-%Y', errors='coerce')
    aqi['date'] = pd.to_datetime(aqi['date'], format='%d-%m-%Y', errors='coerce')

    # --- Define Time Period (Last 3 Years from latest data) ---
    latest_date = idsp['date'].max()
    start_date = latest_date - pd.DateOffset(years=3)

    # Filter both dataframes for the last 3 years
    idsp_filtered = idsp[(idsp['date'] > start_date) & (idsp['date'] <= latest_date)]
    aqi_filtered = aqi[(aqi['date'] > start_date) & (aqi['date'] <= latest_date)]

    # --- Analysis Part 1: Find Top 2 Diseases by State ---
    
    # Group by state and disease, and sum the cases
    disease_cases = idsp_filtered.groupby(['state', 'disease_illness_name'])['cases'].sum().reset_index()
    
    # For each state, find the top 2 diseases
    top_2_diseases = disease_cases.groupby('state').apply(
        lambda x: x.nlargest(2, 'cases')
    ).reset_index(drop=True)

    # --- Analysis Part 2: Find Average AQI by State ---
    
    # Group by state and calculate the average AQI
    avg_aqi_by_state = aqi_filtered.groupby('state')['aqi_value'].mean().round(2).reset_index()
    avg_aqi_by_state.rename(columns={'aqi_value': 'Average AQI (Last 3 Years)'}, inplace=True)

    # --- Combine the Results ---
    
    # Merge the top diseases with the average AQI for each state
    final_report = pd.merge(top_2_diseases, avg_aqi_by_state, on='state', how='left')
    final_report_sorted = final_report.sort_values(by='Average AQI (Last 3 Years)', ascending=False)
    
    # --- Print the Final Report ---
    print(f"--- Top 2 Diseases and Average AQI by State ({start_date.strftime('%d-%m-%Y')} to {latest_date.strftime('%d-%m-%Y')}) ---")
    print(final_report_sorted.to_string(index=False))


except FileNotFoundError:
    print("Error: Make sure 'aqi.csv' and 'idsp.csv' are in the same directory.")
except Exception as e:
    print(f"An error occurred: {e}")
```

7. List the top 5 states with high EV adoption and analyse if their average AQI is significantly better compared to states with lower EV adoption

```python
import pandas as pd

try:
    # Convert date/numeric columns
    aqi['date'] = pd.to_datetime(aqi['date'], format='%d-%m-%Y', errors='coerce')
    aqi['aqi_value'] = pd.to_numeric(aqi['aqi_value'], errors='coerce')

    # --- Part 1: Calculate EV Adoption Rate (2022–2025) ---
    ev_fuel_types = ['ELECTRIC(BOV)', 'PURE EV', 'FUEL CELL HYDROGEN', 'PLUG-IN HYBRID EV', 'STRONG HYBRID EV']

    total_vehicles = vahan.groupby('state')['value'].sum()
    ev_vehicles = vahan[vahan['fuel'].isin(ev_fuel_types)].groupby('state')['value'].sum()

    adoption_df = pd.DataFrame({
        'Total Vehicles': total_vehicles,
        'EVs': ev_vehicles
    }).fillna(0)

    adoption_df['EV Adoption Rate (%)'] = (adoption_df['EVs'] / adoption_df['Total Vehicles'] * 100).round(2)

    # --- Part 2: Clean AQI and Merge ---
    avg_aqi_by_state = aqi.groupby('state')['aqi_value'].mean().round(2)
    adoption_df = adoption_df.merge(avg_aqi_by_state.rename('Avg AQI (Full Period)'), left_index=True, right_index=True, how='left')

    # Remove states with missing AQI
    adoption_df = adoption_df[adoption_df['Avg AQI (Full Period)'].notna()]

    # --- Part 3: Sort and Analyze ---
    adoption_df_sorted = adoption_df.sort_values(by='EV Adoption Rate (%)', ascending=False)
    top_5_states_df = adoption_df_sorted.head(5)
    bottom_5_states_df = adoption_df_sorted.tail(5)

    # Calculate group averages
    top_5_aqi = top_5_states_df['Avg AQI (Full Period)'].mean()
    bottom_5_aqi = bottom_5_states_df['Avg AQI (Full Period)'].mean()

    # --- Part 4: Present Results ---
    print("--- Top 5 States by EV Adoption Rate (2022–2025) ---")
    print(top_5_states_df)

    print("\n--- Bottom 5 States by EV Adoption Rate (2022–2025) ---")
    print(bottom_5_states_df)

    print("\n--- Comparison of Average AQI ---")
    print(f"Average AQI for Top 5 EV States: {top_5_aqi:.2f}")
    print(f"Average AQI for Bottom 5 EV States: {bottom_5_aqi:.2f}")

    print("\n--- Conclusion ---")
    if top_5_aqi < bottom_5_aqi:
        print("✅ States with higher EV adoption have better (lower) average AQI.")
    else:
        print("❌ States with higher EV adoption do not have better AQI.")

except FileNotFoundError:
    print("Error: Make sure 'aqi.csv' and 'vahan.csv' are in the same directory.")
except Exception as e:
    print(f"An error occurred: {e}")
```
